```python
def predict_single_loop(x,w,b):
    n = x.shape[0]
    p = 0
    for i in range(n):
        p_i = x[i]*w[i]
        p += p_i
    
    p += b
    return p 
```

```python
# get a row from our training data
x_vec = X_train[0,:]
print(f"x_vec shape {x_vec.shape}, x_vec value: {x_vec}")

# make a prediction
f_wb = predict_single_loop(x_vec, w_init, b_init)
print(f"f_wb shape {f_wb.shape}, prediction: {f_wb}")
```

```python
def predict(x,w,b):
    return np.dot(x,w)+b
```

```python
def compute_cost(X,y,w,b):
    predictions = np.dot(X,w)+b
    error = np.subtract(y,predictions)**2
    return np.sum(error)/(2*X.shape[0])
```

```python
cost = compute_cost(X_train, y_train, w_init, b_init)
print(f'Cost at optimal w : {cost}')
```

```python
def compute_gradient(X,y,w,b):
    m = X.shape[0]
    no_of_features = X.shape[1]
    gradients = np.zeros(len(w))
    predictions = np.dot(X,w) + b
    error = np.subtract(predictions, y)
    for i in range(no_of_features):
        total = 0
        for j in range(m):
            total += error[j] * X[j][i]
        gradients[i] = total/m 

    db = np.sum(error)/m

    return gradients, db
```

```python
tmp_dj_dw, tmp_dj_db = compute_gradient(X_train, y_train, w_init, b_init)
print(f'dj_db at initial w,b: {tmp_dj_db}')
print(f'dj_dw at initial w,b: \n {tmp_dj_dw}')
```

```python
def gradient_descent(X,y,w_in, b_in, cost_function, gradient_function, alpa, num_iters):
    J_history = []
    w = copy.deepcopy(w_in)
    b = b_in

    for i in range(num_iters):
        dj_dw, dj_db = gradient_function(X,y,w,b)

        w = w - (alpa * dj_dw)
        b = b - (alpa * dj_db)

        if i<100000:      # prevent resource exhaustion 
            J_history.append( cost_function(X, y, w, b))

        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")

    return w, b, J_history
```
